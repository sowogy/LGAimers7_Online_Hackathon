```python
import numpy as np
from scipy.spatial.distance import pdist, squareform

# 데이터 준비 (pivot_t.T = 메뉴별 특징 벡터, 숫자형)
data = pivot_t.T.astype(float)

# 거리 행렬 계산 (cos기준)
dist_matrix = squareform(pdist(data, metric='cosine'))

n_menus = dist_matrix.shape[0]
k = 5  # 가까운 이웃 개수 지정 (적당한 값 선택)

groups = []

for i in range(n_menus):
    # i번째 메뉴와 다른 메뉴들 거리값
    distances = dist_matrix[i]

    # 자기 자신 제외하고 가장 가까운 k개 인덱스 찾기
    # np.argsort는 거리 오름차순 정렬, 첫번째는 자기 자신 거리(0)
    nearest_indices = np.argsort(distances)[1:k+1]

    # i번 메뉴 포함해서 그룹 생성
    group = [i] + nearest_indices.tolist()
    groups.append(group)

# 메뉴명 리스트
menu_names = data.index.tolist()

# 그룹 출력 예시
for idx, group in enumerate(groups):
    group_names = [menu_names[i] for i in group]
    print(f"Group {idx+1} (크기 {len(group_names)}): {group_names}\n")
```

```python
from catboost import CatBoostRegressor, Pool
def smape(y_true, y_pred, eps=1e-8):
    num = np.abs(y_pred - y_true)
    den = (np.abs(y_true) + np.abs(y_pred)).clip(min=eps)
    return 100.0 * np.mean(2.0 * num / den)

def time_holdout_split(n_total, horizon, n_val=None):
    # "조금"만 검증: 전체의 10% 또는 최소 14개 샘플 중 큰 값
    if n_val is None:
        n_val = max(14, n_total // 10)
    n_val = min(n_val, max(1, n_total - (horizon+1)))  # 안전장치
    split = n_total - n_val
    purge = horizon - 1
    split_purged = max(0, split - purge)
    return split_purged, split  # train: [0, split_purged), val: [split, n_total)


# def train_catboost_with_val(X_tr, y_tr, X_val, y_val, cat_features_idx, iterations=500):
#     models = []
#     val_preds = []
#     for d in range(y_tr.shape[1]):
#         w = (y_tr[:, d] != 0).astype(np.float32)  # ← y==0 → weight=0
#         model = CatBoostRegressor(
#             iterations=iterations,
#             depth=6,
#             learning_rate=0.05,
#             loss_function='RMSE',
#             random_seed=42,
#             task_type='GPU',          # CPU면 'CPU'
#             od_type='Iter',           # 조기중단
#             od_wait=50,
#             allow_writing_files=False
#         )
#         model.fit(
#             X_tr, y_tr[:, d],
#             cat_features=cat_features_idx,
#             eval_set=(X_val, y_val[:, d]),
#             use_best_model=True,
#             verbose=False,
#             sample_weight=w
#         )
#         models.append(model)
#         val_preds.append(model.predict(X_val))
#     y_val_hat = np.column_stack(val_preds)
#     return models, y_val_hat
def train_catboost_with_val(X_tr, y_tr, cat_features_idx, iterations=500):
    models = []
    for d in range(y_tr.shape[1]):
        w = (y_tr[:, d] != 0).astype(np.float32)  # ← y==0 → weight=0
        model = CatBoostRegressor(
            iterations=iterations,
            depth=6,
            learning_rate=0.05,
            loss_function='RMSE',
            random_seed=42,
            task_type='CPU',          # CPU면 'CPU'
            od_type='Iter',           # 조기중단
            od_wait=50,
            allow_writing_files=False
        )
        model.fit(
            X_tr, y_tr[:, d],
            cat_features=cat_features_idx,
            verbose=False,
            sample_weight=w
        )
        models.append(model)
    return models


def smape_comp_item(A, P, eps=1e-8):
    """
    공식: (1/T_i) * sum_{t=1..T_i} 2|A-P| / (|A|+|P|)
    단, T_i는 A!=0인 날짜 수.
    return: (score[%], 유효포인트수)
    """
    A = np.asarray(A)
    P = np.asarray(P)
    mask = (A != 0)  # A==0 제외
    n = int(mask.sum())
    if n == 0:
        return np.nan, 0
    num = 2.0 * np.abs(P[mask] - A[mask])
    den = (np.abs(A[mask]) + np.abs(P[mask])).clip(min=eps)
    return 100.0 * np.mean(num / den), n

lookback, horizon = 28, 7
cluster_models = {}
# scores = []
# records = []  
```

```python
# ===========================================================
# - 규칙 준수: 각 샘플의 입력 28일 + 현재 예측 7일만 사용
# - 원예측과 보정값을 섞는 순한맛 보정 + 0 쏠림 방지
# ===========================================================

import pandas as pd
import numpy as np
import glob, os, re
from catboost import Pool

lookback = 28
horizon  = 7

def make_test_input(menus, pivot_test, lookback, pivot_feat=None):
    """
    menus: 학습 시 사용한 메뉴 리스트(열 순서 보존)
    pivot_test: index=영업일자, columns=영업장명_메뉴명, values=매출수량
    pivot_feat: (선택) index=영업일자, 학습 시 사용한 캘린더/도메인 피처 DF
    return: {"__single__": X_input}
    """
    missing = [m for m in menus if m not in pivot_test.columns]
    if missing:
        for m in missing:
            pivot_test[m] = 0.0
    pivot_cluster = pivot_test.reindex(columns=menus).fillna(0.0)

    if len(pivot_cluster) < lookback:
        return {}

    X_lags = pivot_cluster.values[-lookback:, :].ravel()[None, :]  # (1, lookback*M)

    if pivot_feat is not None:
        last_idx = pivot_cluster.index[-1]
        feat_row = pivot_feat.loc[[last_idx]].values  # (1, F)
        X_input = np.empty((1, X_lags.shape[1] + feat_row.shape[1]), dtype=object)
        X_input[0, :X_lags.shape[1]] = X_lags[0]
        X_input[0, X_lags.shape[1]:] = feat_row[0]
    else:
        X_input = X_lags
    return {"__single__": X_input}

# ---------------------- Gentle Calibration (Zero-safe) ----------------------
def _safe_mean(x):
    x = np.asarray(x, float)
    return float(np.nanmean(x)) if x.size else 0.0

def _safe_std(x):
    x = np.asarray(x, float)
    return float(np.nanstd(x, ddof=0)) if x.size else 0.0

def _last_k(arr, k):
    arr = np.asarray(arr, float).reshape(-1)
    k = max(1, min(k, arr.size))
    return arr[-k:]

def _quantile(x, q):
    x = np.asarray(x, float)
    if x.size == 0: return 0.0
    return float(np.nanpercentile(x, q*100.0))

def _transform_base(y, hist_k, mode="quantile"):
    """보정 후보 T(y) 생성: 'scale' | 'meanstd' | 'quantile'"""
    y = np.asarray(y, float).reshape(-1)
    if mode == "scale":
        num = _safe_mean(hist_k); den = _safe_mean(y)
        b = (num/den) if den not in (0.0, np.nan) else 1.0
        a = 0.0
        return a + b*y, {"a": a, "b": b, "mode": "scale"}
    elif mode == "meanstd":
        mu_h, sd_h = _safe_mean(hist_k), _safe_std(hist_k)
        mu_p, sd_p = _safe_mean(y), _safe_std(y)
        if sd_p < 1e-8:
            b = 1.0; a = mu_h - mu_p
        else:
            b = sd_h / sd_p; a = mu_h - b*mu_p
        return a + b*y, {"a": a, "b": b, "mode": "meanstd"}
    else:  # "quantile"
        ql_p, qh_p = _quantile(y, 0.10), _quantile(y, 0.90)
        ql_h, qh_h = _quantile(hist_k, 0.10), _quantile(hist_k, 0.90)
        width_p = max(1e-8, (qh_p - ql_p))
        width_h = (qh_h - ql_h)
        b = width_h / width_p
        mid_p = 0.5*(ql_p + qh_p)
        mid_h = 0.5*(ql_h + qh_h)
        a = mid_h - b*mid_p
        return a + b*y, {"a": a, "b": b, "mode": "quantile"}

def calibrate_gentle_no_leak(
    yhat: np.ndarray,
    history_series: pd.Series | np.ndarray,
    mode: str = "quantile",   # "scale" | "meanstd" | "quantile"
    k: int = 7,
    lam_max: float = 0.25,     # 더 순하게(0.25~0.35 권장)
    gain: float = 0.45,        # λ 민감도(0.4~0.6)
    b_bounds=(0.5, 2.0),       # 과격한 스케일 방지
    range_guard: float | None = None,   # ★ 기본 끔(0 쏠림 방지)
    clip_nonneg: bool = True
):
    # 히스토리(28일)에서 최근 k일만 사용
    if isinstance(history_series, pd.Series):
        hist_full = history_series.values
    else:
        hist_full = np.asarray(history_series, float).reshape(-1)
    hist_k = _last_k(hist_full, k)

    yhat = np.asarray(yhat, float).reshape(-1)
    if yhat.size == 0:
        return yhat, {"lam": 0.0, "mode": mode}

    # === 최근 k일이 거의 0이면 보정 생략(원예측 유지) ===
    if (np.nanmax(hist_k) <= 1e-8) and (np.nanmean(hist_k) <= 1e-8):
        y_final = np.maximum(yhat, 0.0) if clip_nonneg else yhat
        return y_final, {"lam": 0.0, "mode": mode, "reason": "hist_zero"}

    # 1) 보정 후보 T(y)
    Ty, info = _transform_base(yhat, hist_k, mode=mode)
    if "b" in info:
        b = float(np.clip(info["b"], b_bounds[0], b_bounds[1]))
        a = info.get("a", 0.0) + (info.get("b", 1.0) - b) * _safe_mean(yhat)  # 중심 유지 보정
        Ty = a + b*yhat
        info["b"] = b; info["a"] = a

    # 2) λ 계산(표준편차 비율 기반)
    sd_h = _safe_std(hist_k)
    sd_p = _safe_std(yhat)
    if sd_p < 1e-8 and sd_h < 1e-8:
        lam = 0.0
    elif sd_p < 1e-8:
        lam = lam_max
    else:
        r = sd_h / (sd_p + 1e-8)
        lam = np.clip(gain * abs(r - 1.0), 0.0, lam_max)

    y_final = (1.0 - lam) * yhat + lam * Ty

    # 3) 캡은 사용자가 명시했을 때만, cap>0일 때만 적용
    if range_guard is not None:
        cap = float(np.max(hist_k)) * float(range_guard)
        if cap > 0:
            y_final = np.minimum(y_final, cap)

    if clip_nonneg:
        y_final = np.clip(y_final, 0.0, None)

    info.update({"lam": float(lam), "sd_hist": sd_h, "sd_pred": sd_p})
    return y_final, info

# ---------------------- 추론 루프 ----------------------
results = []

# 테스트 파일 경로 (필요 시 경로 조정)
test_files = sorted(glob.glob('test/TEST_*.csv'))

for path in test_files:
    test_df = pd.read_csv(path)
    test_df['영업일자'] = pd.to_datetime(test_df['영업일자'], format='%Y-%m-%d')

    filename    = os.path.basename(path)
    m = re.search(r'(TEST_\d+)', filename)
    test_prefix = m.group(1) if m else os.path.splitext(filename)[0]

    pivot_test = (
        test_df.pivot(index='영업일자', columns='영업장명_메뉴명', values='매출수량')
        .sort_index()
        .fillna(0)
    )

    for target, df in all_results.items():
        if df is None or df.empty:
            continue

        menus = df['메뉴'].tolist()
        # === (A) 입력 구성 ===
        X_test_dict = make_test_input(menus, pivot_test, lookback, pivot_feat)
        if "__single__" not in X_test_dict:
            continue
        X_input = X_test_dict["__single__"]
        if X_input.ndim == 1:
            X_input = X_input.reshape(1, -1)

        pivot_cluster = pivot_test.reindex(columns=menus).fillna(0.0)

        # === (B) 모델 7개로 Raw 예측 ===
        if target not in cluster_models:
            continue
        model_list = cluster_models[target]
        assert len(model_list) == horizon, f"horizon=7이어야 합니다. 현재 {len(model_list)}개."

        _cat_idx = cat_idx if isinstance(cat_idx, list) else []
        pool = Pool(X_input, cat_features=_cat_idx)

        yhat_steps = [m.predict(pool) for m in model_list]   # list of (N,)
        yhat_raw   = np.column_stack(yhat_steps)             # (N, 7)
        yhat_raw_single = yhat_raw[0]                        # (7,)

        # === (C) Gentle 보정 (Zero-safe) ===
        if target not in pivot_cluster.columns:
            continue
        target_hist_full = pivot_cluster[target]
        if len(target_hist_full) < lookback:
            continue
        target_hist_28 = target_hist_full.iloc[-lookback:]  # 반드시 28일만 사용

        yhat_final, calib_info = calibrate_gentle_no_leak(
            yhat=yhat_raw_single,
            history_series=target_hist_28,
            mode="quantile",   # "scale" | "meanstd"로 변경 가능
            k=7,
            lam_max=0.25,
            gain=0.45,
            b_bounds=(0.5, 2.0),
            range_guard=None,  # ★ 끔: 최근 28일 최대가 0이면 전부 0 되는 문제 방지
            clip_nonneg=True
        )
        # 디버그 원하면: print(target, calib_info)

        # === (D) 결과 적재 ===
        for day_offset, val in enumerate(yhat_final, start=1):
            results.append({
                '영업일자': f"{test_prefix}+{day_offset}일",
                '영업장명_메뉴명': target,
                '매출수량': float(val)
            })

# 최종 결과 DataFrame
df_submit = pd.DataFrame(results, columns=['영업일자', '영업장명_메뉴명', '매출수량'])
# 필요 시 저장
# df_submit.to_csv('submission.csv', index=False)
```
